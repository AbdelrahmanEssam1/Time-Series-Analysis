# Time-Series-Analysis

## Importing packages and data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('multiTimeline.csv',skiprows=1)
df.head()
```

```python
df.info()
```

## Wrangle your data

```python
df['Month'] = pd.to_datetime(df.Month)
df.rename(columns={'diet: (Worldwide)':'diet','gym: (Worldwide)':'gym','finance: (Worldwide)':'finance'},inplace=True)
```

```python
df.info()
```

```python
df
```

```python
df.set_index('Month',inplace=True)
```

```python
df.head()
```

```python
df.info()
```

## Exploratory Data Analysis (EDA)

```python
df.plot(figsize=(20,10))
```

```python
df[['diet']].plot(figsize=(20,10))
```

## Trends and Seasonality in Time Series Data

```python
df[['diet']].rolling(12).mean().plot(figsize=(20,10))
```

```python
df[['finance']].plot(figsize=(20,10))
```

```python
df[['finance']].rolling(12).mean().plot(figsize=(20,10))
```

```python
df[['gym']].plot(figsize=(20,10))
```

```python
df[['gym']].rolling(12).mean().plot(figsize=(20,10))
```

```python
pd.concat([df[['diet']].rolling(12).mean(),df[['gym']].rolling(12).mean()],axis=1).plot(figsize=(20,10))
```
